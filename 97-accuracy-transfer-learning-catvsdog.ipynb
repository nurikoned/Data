         Creating a 10% Subset and Splitting Data into Train and Test Sets
    </p>
    <p style="font-size: 1.0em; font-family: 'Arial', sans-serif; text-align: justify; line-height: 1.5; padding: 10px;">
        In this section, we prepare a manageable subset of our dataset by taking a random 10% sample of the available images for both classes (Cats and Dogs). We then split this subset further into training (80%) and testing (20%) sets to facilitate efficient model development. This step includes creating the necessary directories, selecting a subset of images, and moving these images into respective folders for training and testing. This streamlined approach enables quicker training and testing cycles, especially when working with limited computational resources.
    </p>
</div>


```python
import os
import shutil
from pathlib import Path
from sklearn.model_selection import train_test_split

# Define paths
data_dir = Path('/kaggle/input/microsoft-catsvsdogs-dataset/PetImages/')
train_dir = Path('train')
test_dir = Path('test')

# Create directories
for folder in ['Cat', 'Dog']:
    os.makedirs(train_dir / folder, exist_ok=True)
    os.makedirs(test_dir / folder, exist_ok=True)

# Get image file paths
cat_images = [file for file in (data_dir / 'Cat').glob('*.jpg') if file.is_file()]
dog_images = [file for file in (data_dir / 'Dog').glob('*.jpg') if file.is_file()]

# Take 10% of the data
cat_subset, _ = train_test_split(cat_images, test_size=0.9, random_state=42)
dog_subset, _ = train_test_split(dog_images, test_size=0.9, random_state=42)

# Split the 10% data into training and test sets (80% train, 20% test)
cat_train, cat_test = train_test_split(cat_subset, test_size=0.2, random_state=42)
dog_train, dog_test = train_test_split(dog_subset, test_size=0.2, random_state=42)

# Function to move files
def move_files(files, target_dir):
    for file_path in files:
        try:
            shutil.copy(file_path, target_dir / file_path.name) 
        except Exception as e:
            print(f"Error copying file {file_path}: {e}")

# Move files to respective directories
move_files(cat_train, train_dir / 'Cat')
move_files(cat_test, test_dir / 'Cat')
move_files(dog_train, train_dir / 'Dog')
move_files(dog_test, test_dir / 'Dog')

print("Data successfully divided into training and testing sets.")
```

<div style="text-align: center;">
    <p style="background-color: #8aeb92; color: #910909; font-size: 2.0em; font-family: 'Arial', sans-serif; text-align: center; border-radius: 25px; border: 4px solid #910909; display: inline-block; padding: 20px 20px; margin: 0 auto; line-height: 1.5; font-weight: bold; box-shadow: 0px 4px 10px rgba(0, 0, 0, 0.3);">
         Function to validate and clean image files
    </p>
    <p style="font-size: 1.0em; font-family: 'Arial', sans-serif; text-align: justify; line-height: 1.5; padding: 10px;">
        In this step, we ensure data quality by validating each image file in the dataset. The function checks every file to confirm it is a valid image, and any corrupted or unreadable files are automatically removed. This process helps prevent issues during model training that may arise from invalid or corrupted images, ensuring a cleaner, more reliable dataset. Cleaning the data in this way supports the stability and performance of our model.
    </p>
</div>


```python
import os
from PIL import Image

def validate_and_remove_images(directory):
    for subdir, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(subdir, file)
            try:
                img = Image.open(file_path)
                img.verify()  # Verify that it is an image
            except (IOError, SyntaxError) as e:
                print(f"Removing bad file: {file_path}")
                os.remove(file_path)
```

<div style="text-align: center;">
    <p style="background-color: #8aeb92; color: #910909; font-size: 2.0em; font-family: 'Arial', sans-serif; text-align: center; border-radius: 25px; border: 4px solid #910909; display: inline-block; padding: 20px 20px; margin: 0 auto; line-height: 1.5; font-weight: bold; box-shadow: 0px 4px 10px rgba(0, 0, 0, 0.3);">
         Preparing Data Generators for Training and Testing
    </p>
    <p style="font-size: 1.0em; font-family: 'Arial', sans-serif; text-align: justify; line-height: 1.5; padding: 10px;">
        In this step, we set up data generators to feed images into the model for training and validation. Using TensorFlow's <code>ImageDataGenerator</code>, we preprocess the images by rescaling their pixel values. First, we clean the training and testing directories by removing any invalid images, ensuring a smooth data flow. The data generators are then configured to load images from these directories, resize them to 224x224 pixels, and set a batch size of 32. This setup streamlines the data pipeline, preparing it for efficient model training and evaluation.
    </p>
</div>


```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from pathlib import Path 

tf.random.set_seed(42)

train_datagen = ImageDataGenerator(rescale = 1./255)
valid_datagen = ImageDataGenerator(rescale = 1./255)


train_dir = Path('train')
test_dir = Path('test')  

validate_and_remove_images('train')
validate_and_remove_images('test')

train_data = train_datagen.flow_from_directory(train_dir,
                                               target_size = (224, 224),
                                               batch_size = 64, class_mode = 'binary', seed = 42)
test_data = valid_datagen.flow_from_directory(test_dir,
                                               target_size = (224, 224),
                                               batch_size = 64, class_mode = 'binary', seed = 42)
```

    <p style="background-color: #8aeb92; color: #910909; font-size: 2.0em; font-family: 'Arial', sans-serif; text-align: center; border-radius: 25px; border: 4px solid #910909; display: inline-block; padding: 20px 20px; margin: 0 auto; line-height: 1.5; font-weight: bold; box-shadow: 0px 4px 10px rgba(0, 0, 0, 0.3);">

        In this project, we utilize pre-trained models to apply transfer learning, leveraging their powerful feature extraction capabilities.
```python
resnet_url = "https://www.kaggle.com/models/tensorflow/resnet-50/TensorFlow2/classification/1"
efficientnet_url = "https://www.kaggle.com/models/google/efficientnet-v2/TensorFlow2/imagenet1k-b0-classification/2"
mobilenet_url = "https://kaggle.com/models/google/mobilenet-v2/TensorFlow2/100-224-feature-vector/1"
```

<div style="text-align: center;">
    <p style="background-color: #8aeb92; color: #910909; font-size: 2.0em; font-family: 'Arial', sans-serif; text-align: center; border-radius: 25px; border: 4px solid #910909; display: inline-block; padding: 20px 20px; margin: 0 auto; line-height: 1.5; font-weight: bold; box-shadow: 0px 4px 10px rgba(0, 0, 0, 0.3);">
         Building a Transfer Learning Model
    </p>
    <p style="font-size: 1.0em; font-family: 'Arial', sans-serif; text-align: justify; line-height: 1.5; padding: 10px;">
        Here, we define a function to create a transfer learning model using TensorFlow Hub. This function, <code>create_model</code>, accepts a URL for a pre-trained model and the number of output classes. It builds a model with two main components: a pre-trained feature extraction layer that captures high-level patterns from the input images, and a dense output layer for binary classification. By freezing the feature extraction layer, we focus on training only the output layer, which helps speed up the training process and reduce the risk of overfitting. This approach efficiently adapts a powerful pre-trained model for our specific classification task.
    </p>
</div>


```python
from tensorflow.keras import layers
import tensorflow_hub as hub
import tf_keras as keras
IMAGE_SHAPE = (224,224)
def create_model(model_url, num_classes):
 
  feature_extractor_layer = hub.KerasLayer(model_url,
                                           trainable=False,
                                           name='feature_extraction_layer',
                                           input_shape=IMAGE_SHAPE+(3,)) 

 
  model = keras.Sequential([
    feature_extractor_layer, 
    keras.layers.Dense(num_classes, activation='sigmoid', name='output_layer') 
      
  ])
  return model
```

<div style="text-align: center;">
    <p style="background-color: #8aeb92; color: #910909; font-size: 2.0em; font-family: 'Arial', sans-serif; text-align: center; border-radius: 25px; border: 4px solid #910909; display: inline-block; padding: 20px 20px; margin: 0 auto; line-height: 1.5; font-weight: bold; box-shadow: 0px 4px 10px rgba(0, 0, 0, 0.3);">
         Compiling and Training the ResNet Model
    </p>
    <p style="font-size: 1.0em; font-family: 'Arial', sans-serif; text-align: justify; line-height: 1.5; padding: 10px;">
        In this step, we compile and train our ResNet model for binary classification. We start by creating an instance of the model with the pre-trained ResNet-50 feature extraction layer, designed for efficient image feature capture. The model is compiled with binary cross-entropy loss, suitable for two-class classification problems, and the Adam optimizer, which helps with fast and reliable convergence. The model is then trained on our prepared training data for 5 epochs, providing a strong foundation for our binary classification task. This setup allows the model to leverage ResNets robust feature extraction while focusing on fine-tuning the classification layer for optimal accuracy.
    </p>
</div>


```python
resnet_model = create_model(resnet_url, num_classes=1)

resnet_model.compile(loss='binary_crossentropy',
                     optimizer=keras.optimizers.Adam(),
                     metrics=['accuracy'])

resnet_history = resnet_model.fit(train_data,
                                  epochs=5,
                                  steps_per_epoch=len(train_data),
                                  #validation_data=test_data,
                                  #validation_steps=len(test_data),
                                  verbose = 1) 
```

         Compiling and Training the EfficientNet Model
    </p>
    <p style="font-size: 1.0em; font-family: 'Arial', sans-serif; text-align: justify; line-height: 1.5; padding: 10px;">
        In this step, we compile and train an EfficientNet model for our binary classification task. Using the <code>create_model</code> function, we initialize a model based on the EfficientNet architecture, which is known for its balance of high accuracy and computational efficiency. The model is compiled with binary cross-entropy loss, a suitable choice for binary classification, and the Adam optimizer, which enhances training efficiency. Training the model for 5 epochs on the prepared dataset allows us to fine-tune EfficientNets pre-trained feature extraction layer for our specific dataset, aiming for accurate and reliable classification results.

```python
efficientnet_model = create_model(efficientnet_url, num_classes = 1)

efficientnet_model.compile(loss='binary_crossentropy',
                     optimizer=keras.optimizers.Adam(),
                     metrics=['accuracy'])

efficientnet_history = efficientnet_model.fit(train_data,
                                  epochs=5,
                                  steps_per_epoch=len(train_data),
                                  #validation_data=test_data,
                                  #validation_steps=len(test_data),
                                  verbose = 1) 
```

```python
mobilenet_model = create_model(mobilenet_url, num_classes = 1)

mobilenet_model.compile(loss='binary_crossentropy',
                     optimizer=keras.optimizers.Adam(),
                     metrics=['accuracy'])

mobilenet_history = mobilenet_model.fit(train_data,
                                  epochs=5,
                                  steps_per_epoch=len(train_data),
                                  #validation_data=test_data,
                                  #validation_steps=len(test_data),
                                  verbose = 1) 
```

         Evaluating Model Performance on Test Data
    </p>
    <p style="font-size: 1.0em; font-family: 'Arial', sans-serif; text-align: justify; line-height: 1.5; padding: 10px;">
        In this section, we evaluate the performance of our three models (ResNet, MobileNet, and EfficientNet) on the test dataset. We calculate the total number of test samples and the required steps for evaluation to ensure that all data is included. Each model is then evaluated on the test data, and we retrieve the accuracy scores for comparison. This step allows us to assess how well each pre-trained model has adapted to our binary classification task, helping us choose the most accurate model for our use case.
    </p>
</div>


```python
# Number of samples
num_test_samples = test_data.samples
print(f"Total number of test samples: {num_test_samples}")

# Calculate steps, ensuring all data is included
batch_size = test_data.batch_size
steps = (num_test_samples // batch_size) + int(num_test_samples % batch_size != 0)
print(f"Total steps for evaluation: {steps}")

# Evaluate the model
resnet_results = resnet_model.evaluate(test_data, steps=steps)
mobilenet_results = mobilenet_model.evaluate(test_data, steps=steps) 
efficientnet_results = efficientnet_model.evaluate(test_data, steps=steps)

resnet_accuracy = resnet_results[1] 
mobilenet_accuracy = mobilenet_results[1] 
efficientnet_accuracy = efficientnet_results[1] 

print(f"Resnet accuracy: {resnet_accuracy:.4f}")
print(f"MobileNet accuracy: {mobilenet_accuracy:.4f}") 
print(f"EfficientNet accuracy: {efficientnet_accuracy:.4f}")
```
